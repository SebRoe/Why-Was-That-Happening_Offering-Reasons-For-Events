```python
def show_all_timelines(rollout):
    # Plot each column of the rollout as a timeline (one timeline per column)
    df = pd.DataFrame(rollout)
    columns = df.columns

    fig, ax = plt.subplots(len(columns)//2 + 1, 2, figsize=(50, 30))
    axes = ax.flatten() 
    for axs, col in zip(axes, columns):
        sns.lineplot(data=df[col], ax=axs)
        axs.set_title(col)
        
```
